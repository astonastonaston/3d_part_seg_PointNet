```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from csv import DictReader
```

# Average Training Loss and Average Training Accuracy

```python
train_data = pd.read_csv('training.csv')
training_loss= train_data['train_loss']
epoch= train_data['epoch']
training_accuracy= train_data['accuracy']
```

```python
avg_error= []
for i in range(120):
    sum=0
    for a in range(len(training_loss)):
        if(i == epoch[a]):
            sum+= training_loss[a]
        else:
            pass
    avg= sum/380
    avg_error.append(avg)
```

```python
print(len(avg_error))
```

```python
avg_train_accuracy=[]
for i in range(120):
    sum=0
    for a in range(len(training_accuracy)):
        if(i == epoch[a]):
            sum+= training_accuracy[a]
        else:
            pass
    avg= sum/380
    avg_train_accuracy.append(avg)
```

```python
print((avg_train_accuracy))
```

# Average Validation Error and Average Validation Accuracy

```python
validation_data= pd.read_csv('validation.csv')
validation_loss= validation_data['validation_loss']
epoch_val= validation_data['epoch']
```

```python
len(validation_loss)
```

```python
avg_validation_error= []
for i in range(120):
    sum=0
    for a in range(len(validation_loss)):
        if(i == epoch_val[a]):
            sum+= validation_loss[a]
        else:
            pass
    avg_val= sum/38
    avg_validation_error.append(avg_val)
```

```python
print(avg_validation_error)
```

```python
validation_accuracy= validation_data['accuracy']
len(validation_accuracy)
```

```python
avg_validation_accuracy=[]
for i in range(120):
    sum=0
    for a in range(len(validation_accuracy)):
        if(i == epoch_val[a]):
            sum+= validation_accuracy[a]
        else:
            pass
    avg_acc= sum/38
    avg_validation_accuracy.append(avg_acc)
```

```python
print(len(avg_validation_accuracy))
```

# Plotting

```python
plt.xlabel('Epoch')
plt.ylabel('Error')
#plt.title('')
plt.plot(avg_error, label= 'training_Error')
plt.plot(avg_validation_error, label= 'Validation_Error')
plt.legend()
plt.grid()
```

```python
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.plot(avg_train_accuracy,label= 'training_Accuracy')
plt.plot(avg_validation_accuracy,label= 'Validation_Accuracy')
plt.legend()
plt.grid()
```
